```python
import geopandas as gpd
import pandas as pd
import numpy as np

df_manzanas = gpd.read_file('predata/manzanas_medellin.geojson')

df_copy = df_manzanas.copy()

df_copy = df_copy.to_crs(epsg=32618)

df_copy['area'] = df_copy.geometry.area
df_manzanas['area'] = df_copy['area']

df_barrios = gpd.read_file('predata/barrios.geojson')
df_barrios = df_barrios.to_crs(epsg=4326)
df_manzanas = df_manzanas.to_crs(epsg=4326)

df_usos = gpd.read_file('predata/usos.geojson')
joined = gpd.sjoin(df_manzanas, df_barrios, how="inner", op="intersects")
joined
```

```python
joined.reset_index(drop=True, inplace=True)
joined.drop(columns=['index_right'], inplace=True)
```

```python
df_final = joined
```

```python
df_usos = df_usos.to_crs(epsg=4326)
```

```python
df_final = gpd.sjoin(df_final, df_usos, how="inner", op="intersects")
```

```python
df_final=df_final[df_final['SUBCATEGORIA'] == 'Áreas predominantemente residenciales']
df_final.drop(columns=['index_right'], inplace=True)
```

```python
df_final
```

```python
df_estratos = gpd.read_file('predata/estrato.geojson')
df_estratos = df_estratos.dropna(subset=['MANZANA'])
df_estratos.reset_index(drop=True, inplace=True)
df_estratos = df_estratos.to_crs(epsg=4326)
df_promedio_estrato = df_estratos.groupby('MANZANA')['ESTRATO'].mean().reset_index()

df_promedio_estrato['ESTRATO_PROMEDIO'] = np.floor(df_promedio_estrato['ESTRATO'])

df_estratos = df_estratos.merge(df_promedio_estrato[['MANZANA', 'ESTRATO_PROMEDIO']], on='MANZANA', how='left')

df_estratos = df_estratos.drop_duplicates(subset='MANZANA')

df_estratos.drop(columns=['ESTRATO'], inplace=True)

df_estratos.rename(columns={'ESTRATO_PROMEDIO': 'ESTRATO'}, inplace=True)
```

```python
df_estratos.reset_index(drop=True, inplace=True)
```

```python
df_final1 = gpd.sjoin(df_final, df_estratos, how="left", op="intersects")
```

```python
cols=["geometry", "area", "MANZANA", "NOMBRE", "BARRIO", "ESTRATO"]
df_final1=df_final1[cols]
```

```python
df_final1=df_final1.dropna()
df_final1 = df_final1.drop_duplicates(subset='geometry')
```

```python
df_final1.reset_index(drop=True, inplace=True)
df_final=df_final1
```

```python
df_poblacion_barrios = pd.read_csv('predata/poblacion-barrios.csv')

df_area_barrio = df_final.groupby('NOMBRE')['area'].sum().reset_index()
df_area_barrio.columns = ['NOMBRE', 'AREA_TOTAL_BARRIO']

df_final = pd.merge(df_final, df_area_barrio, on='NOMBRE', how='left')
df_final = pd.merge(df_final, df_poblacion_barrios, on='NOMBRE', how='left')
df_final['POBLACION_MANZANA'] = df_final['POBLACION'] * df_final['area']/df_final['AREA_TOTAL_BARRIO']

df_final['POBLACION_MANZANA'].fillna(0, inplace=True)
df_final['POBLACION_MANZANA'] = np.ceil(df_final['POBLACION_MANZANA'])
df_final
```

```python
df_final.to_file("data.geojson", driver='GeoJSON')
```

```python
import folium
mapa = folium.Map(location=[df_final.geometry.centroid.y.mean(), df_final.geometry.centroid.x.mean()], zoom_start=13)

for idx, row in df_final.iterrows():
    area_formateada = "{:.2f}".format(row['area'])
    etiqueta = f"Barrio: {row['NOMBRE']}, Área: {area_formateada} m², Población: {row['POBLACION_MANZANA']}, estrato: {row['ESTRATO']}"
    folium.GeoJson(row.geometry, tooltip=etiqueta).add_to(mapa)

mapa.save('mapa.html')
```
